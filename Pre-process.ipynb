```python
import requests

# URL for the Agora dataset zip file (update this with the actual download link)
dataset_url = "https://zenodo.org/records/14670665/files/agora.zip?download=1"

print("Downloading dataset...")
response = requests.get(dataset_url)
if response.status_code == 200:
    print("Download complete!")
else:
    print(f"Download failed with status code {response.status_code}")
    exit()

# Save the downloaded content as a zip file locally
zip_filename = "agora_dataset.zip"
with open(zip_filename, "wb") as f:
    f.write(response.content)

print(f"Saved dataset to {zip_filename}")
```

```python
import zipfile
import os

# Define the extraction directory
extract_dir = "agora_dataset"

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Extract the downloaded zip file into the extraction directory
with zipfile.ZipFile(zip_filename, "r") as zip_ref:
    zip_ref.extractall(extract_dir)

print("Extraction complete. Extracted files:")
print(os.listdir(extract_dir))
```

```python
import pandas as pd

# Suppose the extracted folder contains a CSV file (for example, 'documents.csv')
file_name = "data/documents.csv"
csv_path = os.path.join(extract_dir, file_name)

if os.path.exists(csv_path):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_path)

    # View the first few rows of the dataset
    print("First 5 rows of the CSV file:")
    print(df.head(20))

    # Also, display DataFrame info such as data types and non-null counts
    print("\nDataset information:")
    df.info()
else:
    print(f"CSV file 'documents.csv' not found in the folder {extract_dir}.")
```
